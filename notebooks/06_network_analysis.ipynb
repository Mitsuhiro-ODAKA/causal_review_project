```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tw = pd.read_csv("C:/Users/user/OneDrive/Desktop/causal_review_project/data/processed/causal_extraction_edge_list_network_analysis.csv")
tw.head()
```

```python
tw01 = tw.sort_values("BetweennessCentrality", ascending=False)[0:10]

plt.rcParams['font.size']=20
plt.rcParams['font.family']='Yu Gothic'
plt.figure(figsize=(10, 6))  # 図のサイズは任意で調整
plt.barh(tw01['name'], tw01['BetweennessCentrality'], color='darkblue')  # 棒の色も調整可能
plt.xlabel('ノード名')
plt.ylabel('媒介中心性')
# plt.title('Betweenness Centrality by Node')
plt.xticks(rotation=45)  # ラベルが重なるときは回転させる
plt.gca().invert_yaxis()  # 上に大きい値が来るように順序を反転
plt.tight_layout()  # レイアウトの調整
plt.savefig("C:/Users/user/OneDrive/Desktop/causal_review_project/imgs/BetweennessCentrality.png")
plt.show()
```
